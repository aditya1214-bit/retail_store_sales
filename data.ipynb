```python
import pandas as pd 
```

```python
df = pd.read_csv("retail_store_sales.csv")
```

```python
#cleaning the columns names

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
)
```

```python
#fixing date columnns 
df["transaction_date"] = pd.to_datetime(
    df["transaction_date"],
    errors = "coerce"
)
```

```python
df
```

```python
df
```

```python
#for capitalising 
data_new = []

for i in df.columns:
    data_new.append(i.capitalize())
```

```python
df.columns = data_new
```

```python
df
```

```python
#handle missing values 
#for missing items names 
df["Item"] = df["Item"].fillna("UNKNOWN ITEM")

#for discount columns 
df["Discount_applied"] = df["Discount_applied"].fillna(False)

#for changing types 
df["Discount_applied"] = df["Discount_applied"].astype(bool)
```

```python
df
```

```python
#fixx numerical columns 
numeric_col = ["Price_per_unit","Quantity", "Total_spent"]

for col in numeric_col:
    df[col] = pd.to_numeric(df[col], errors = "coerce")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(
    df['Total_spent'],
    bins=30,
    kde=True
)

plt.title("Distribution of total spent ")
plt.xlabel("Total spent")
plt.ylabel("Frequency")

plt.show()
```

```python
#Categorywise sales
sns.barplot(
    x='Category',
    y='Total_spent',
    data=df,
    estimator=sum
)

plt.xticks(rotation=45)

plt.title("Category wise Total sales ")
plt.xlabel("Category")
plt.ylabel("Total sales")

plt.show()
```

```python
#payment method analysis
sns.countplot(
    x='Payment_method',
    data = df
)

plt.title("Payment method usage")
plt.xlabel("payment method")
plt.ylabel("Count")

plt.show()
```

```python
#location wise revenue 
sns.barplot(
    x='Location',
    y='Total_spent',
    data=df,
    estimator=sum
)

plt.xticks(rotation=45)

plt.title("Revenue by location")
plt.xlabel("location")
plt.ylabel("Revenue")

plt.show()
```

```python
#saatter plot - quantity vs total spent 
sns.scatterplot(
    x='Quantity',
    y='Total_spent',
    hue='Category',
    data=df
)

plt.title("Quantity vs Total spent ")

plt.show()
```

```python
#discount impact Analysis 
sns.boxplot(
    x='Discount_applied',
    y='Total_spent',
    data=df
)

plt.title("Discount impact on spending")
plt.show()
```

```python
#correlation heatmap

numeric_df = df.select_dtypes(include=['float','int64'])

corr =numeric_df.corr()

sns.heatmap(
    corr,
    annot=True,
    cmap = 'coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
#daily sales trend
daily_sales = df.groupby(
    'Transaction_date'
)['Total_spent'].sum().reset_index()

sns.lineplot(
    x='Transaction_date',
    y='Total_spent',
    data = daily_sales
)

plt.title("Daily sales trends")
plt.xticks(rotation=45)
plt.show()
```

```python
#top 10 selling items 
top_items = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False).head(10)

sns.barplot(
    x=top_items.values,
    y= top_items.index
)

plt.title("top 10 selling items ")
plt.xlabel("Quantity sold")
plt.ylabel("Items")

plt.show()
```

```python
#pairplot
sns.pairplot(
    df[['Price_per_unit', 'Quantity','Total_spent']]
)
plt.show()
```

```python
#monthly sales analysis 
#exttract month
df['Month'] = df['Transaction_date'].dt.month

monthly_sales = df.groupby('Month')['Total_spent'].sum().reset_index()

sns.lineplot(
    x='Month',
    y='Total_spent',
    data = monthly_sales,
    marker='o'
)

plt.title("Monthly sales Trend")
plt.show()
```

```python
#categoru count
sns.countplot(
    y='Category',
    data = df,
    order = df['Category'].value_counts().index
)

plt.title("Number of products per category")
plt.show()
```

```python
#customer purchase analysis 
customer_spending = df.groupby(
    'Customer_id'
)['Total_spent'].sum().sort_values(ascending=False).head(10)

sns.barplot(
    x = customer_spending.values,
    y = customer_spending.index
)

plt.title("Top 10 Customer by spending ")

plt.xlabel("Total spending")
plt.ylabel("Customer ID")

plt.show()
```
